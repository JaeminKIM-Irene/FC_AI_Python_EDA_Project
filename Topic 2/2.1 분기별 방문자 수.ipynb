```python
# 2) 선택국가의 데이터(연,월,방문객수,연간증가율) 추출과
#    결측치를 처리한 데이터프레임 생성
#------------------
selectedCountryKor = '일본'
japanDataFrame = ut2.getCountryDataFrame(asiaDataFrame,selectedCountryKor)
japanDataFrame
```

```python
# 3) 데이터 타입변환을 위한 info출력
#------------------
japanDataFrame.info()
```

```python
# 4) 연속된 값으로 날짜처리(Date는 DataFrame의 index로 실제 csv에 반영X)
#------------------
japanDataFrame = ut2.changeYearAndMonthDtypeToInt(japanDataFrame)
japanDataFrame
```

```python
# 5) 선택한 날짜로 필터링
#------------------
startDate = '2018-06-01'
endDate = '2019-05-01'
selectedData = ut2.extractDataByDateFilter(japanDataFrame,startDate,endDate)
display(selectedData)
```

```python
quarter = []
selected = selectedData.values
cnt = 1

for idx in range(int(len(selected)/3)) :
    i = idx * 3
    mean_visitor = np.mean([selected[i][2], selected[i+1][2], selected[i+2][2]])
    mean_rate = np.mean([selected[i][3], selected[i+1][3], selected[i+2][3]])
    quarter.append([selected[i][0], selected[i][1], mean_visitor, mean_rate, str(cnt)+'분기'])
    
    if cnt < 4 : cnt += 1
    else : cnt = 1
    
quarter = pd.DataFrame(quarter,
                       columns=['Year', 'Month', 'NumVisitorsAv', 'annualGrowthRateAv', 'quarter'])
```

```python
# 6.xx) Year 와 Month를 object로 변환(선택)
#------------------
arrangedQuarter = ut2.changeYearAndMonthDtypeToObject(quarter)
arrangedQuarter['yearQuarter'] = arrangedQuarter['Year'].astype(str) + ' ' + arrangedQuarter['quarter'].astype(str)
arrangedQuarter
```

```python
# 6.xx) 날짜 필터링된 데이터프레임의 시작과 끝 년월을 문자열로 추출
#------------------
startYear,startMonth,endYear,endMonth = ut2.getStrYearAndMonth(arrangedQuarter)
```

```python
# 7) 결과를 CSV로 저장
#------------------
countryNameEng = 'japan'
fileName = f'./{countryNameEng}DataFrame ({startYear}{startMonth}~{endYear}{endMonth})_quarter.csv'
arrangedQuarter.to_csv(fileName, encoding='utf-8-sig', index=False)
```
